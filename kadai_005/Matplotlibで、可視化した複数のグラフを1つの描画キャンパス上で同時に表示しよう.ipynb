```python
#matplotlib.pyplotメソッドとnumpyをインポートするコードを記述する。
import matplotlib.pyplot as plt
import numpy as np

#Figure(描画領域全体)とAxes(ひとつひとつのプロットを描く領域)を作成するコードを記述する。
fig, axes = plt.subplots(2, 3, tight_layout=True)

#各領域に、折れ線グラフ、sin関数、ヒストグラム、散布図のデータを表示するコードを記述する。
#折れ線グラフ
y0 = [1, 2, -5, 2]
axes[0,0].plot(y0)

#sin関数
x1 = np.linspace(0, 10, 30)
y1 = 2 + 2 * np.sin(2 * x1)
axes[0,1].plot(y1)

#ヒストグラム
y2 = np.random .normal(15, 5, 2000)
axes[0,2].hist(y2)

#散布図
np.random.seed(3)
x3 = 4 + np.random.normal(0, 2, 60)
y3 = 4 + np.random.normal(0, 2, len(x3))
axes[1,0].scatter(x3, y3)

#グラフを表示させる行「plt.show()」を書く。
plt.show()
```
